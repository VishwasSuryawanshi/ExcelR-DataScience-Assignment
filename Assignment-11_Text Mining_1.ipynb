## Q1) Perform sentimental analysis on the Elon-musk tweets (Exlon-musk.csv)

```python
import numpy as np
import pandas as pd
import string
import spacy

from matplotlib.pyplot import imread
from matplotlib import pyplot as plt
from wordcloud import WordCloud
%matplotlib inline
```

```python
data = pd.read_csv("Elon_musk.csv",error_bad_lines=False)
data.iloc[:,1:2]
```

```python
data = [Text.strip() for Text in data.Text]
data = [Text for Text in data if Text]
data[0:20]
```

```python
# Part of Speech Tagging
nlp = spacy.load("en_core_web_sm")

one_data = data[8]
doc_data = nlp(one_data)
spacy.displacy.render(doc_data, style="ent", jupyter=True)
```

```python
for token in doc_data:
    print(token, token.pos_)
```

```python
# Filtering Noun and verbs
nouns_verbs = [token.text for token in doc_data if token.pos_ in ("NOUN","VERB")]
print(nouns_verbs[0:10])
```

```python
# Counting Tokens again
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()

X = cv.fit_transform(nouns_verbs)
sum_words = X.sum(axis=0)
words_freq = [(word, sum_words[0, idx]) for word, idx in cv.vocabulary_.items()]
words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
wf_df = pd.DataFrame(words_freq)
wf_df.columns = ['word', 'count']

wf_df[0:10]
```

```python
##Visualizing results
#Barchart for top 10 nouns + verbs
wf_df[0:10].plot.bar(x='word', figsize=(12,8), title='Top verbs and nouns')
```

```python
import numpy as np
import pandas as pd 
import string 
import spacy 

from matplotlib.pyplot import imread
from matplotlib import pyplot as plt
from wordcloud import WordCloud
%matplotlib inline


data = pd.read_csv("Elon_musk.csv",error_bad_lines=False)
data = [Text.strip() for Text in data.Text] 
data = [Text for Text in data if Text] 
```

```python
from nltk import tokenize
sentences = tokenize.sent_tokenize(" ".join(data))
sentences[5:15]
```

```python
sent_df = pd.DataFrame(sentences, columns=['sentence'])
sent_df
```

```python
affinity_scores = afinn.set_index('word')['value'].to_dict()
affinity_scores
```

```python
nlp = spacy.load('en_core_web_sm')
sentiment_lexicon = affinity_scores

def calculate_sentiment(Text: str = None):
    sent_score = 0
    if Text:
        sentence = nlp(Text)
        for word in sentence:
            sent_score += sentiment_lexicon.get(word.lemma_, 0)
    return sent_score
```

```python
calculate_sentiment(Text = "amazing")
```

```python
sent_df['sentiment_value'] = sent_df['sentence'].apply(calculate_sentiment)
sent_df['word_count'] = sent_df['sentence'].str.split().apply(len)
sent_df['word_count'].head(10)
```

```python
sent_df.sort_values(by="sentiment_value").tail(10)
```
