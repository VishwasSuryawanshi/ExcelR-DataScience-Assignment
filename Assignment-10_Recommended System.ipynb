## Build a recommender system by using cosine similarities score.

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("C:\\Users\\Lenovo\\Desktop\\ExcelR\\Data Science Assignments\\10_Recommendation Systems\\book.csv")
data
```

```python
len(data.BookTitle.unique())
```

```python
len(data.UserID.unique())
```

```python
data1 = data.pivot_table(index="UserID", columns="BookTitle", values="BookRating", aggfunc="sum")
data1
```

```python
data1.index = data.UserID.unique()
data1
```

```python
data1.fillna(0,inplace=True)
data1
```

```python
# Calculating Cosine Similarity between users :
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
data_user = 1-pairwise_distances(data1.values,metric="cosine")
data_user
```

```python
data_user_df = pd.DataFrame(data_user)
```

```python
data_user_df.index = data.UserID.unique()
data_user_df.columns = data.UserID.unique()
```

```python
data_user_df.iloc[0:5,0:5]
```

```python
np.fill_diagonal(data_user,0)
data_user_df.iloc[0:5,0:5]
```

```python
# Most similar users :
data_user_df.idxmax(axis=1)
```
